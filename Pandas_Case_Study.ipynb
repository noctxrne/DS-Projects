```python
import pandas as pd
import numpy as np
```

```python
url="https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv"
df=pd.read_csv(url)
print(f"Head is \n{df.head()}")
```

```python
print(f"Shape is {df.shape}")
df.info()
```

```python
print("Data Types:\n", df.dtypes)
```

```python
print("Missing Values:\n", df.isna().sum())
```

```python
df = df.drop_duplicates()
df
```

```python
df.rename(columns={'Title': 'Movie_Title'}, inplace=True)
df
```

```python
unique_genres=df['Genre'].unique()
print("\nUnique Genres:",unique_genres)
```

```python
df['Genre_Count'] = df['Genre'].apply(lambda x: len(x.split(',')))
df
```

```python
# 8. Handle missing values in 'Revenue (Millions)' and 'Metascore'
df[['Revenue (Millions)','Metascore']].isnull().sum()
```

```python
# 9. Top 5 movies by Rating
top_5_rated = df.sort_values(by='Rating', ascending=False).head(5)
print("Top 5 Movies by Rating:\n", top_5_rated[['Movie_Title', 'Rating']])
```

```python
# 10. Movies with highest/lowest revenue
max_revenue = df[df['Revenue (Millions)'] == df['Revenue (Millions)'].max()]
min_revenue = df[df['Revenue (Millions)'] == df['Revenue (Millions)'].min()]
print("Highest Revenue Movie:\n", max_revenue[['Movie_Title', 'Revenue (Millions)']])
print("Lowest Revenue Movie:\n", min_revenue[['Movie_Title', 'Revenue (Millions)']])
```

```python
# 11. Average rating by Director
avg_rating_director = df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
print("Average Rating by Director:\n", avg_rating_director)
```

```python
# 12. Movies per year
movies_per_year = df['Year'].value_counts().sort_index()
print("Movies per Year:\n", movies_per_year)
```

```python
# 13. Movies with rating > 8 and revenue > 100M
high_rating_high_revenue = df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)]
print("High Rating & High Revenue Movies:\n", high_rating_high_revenue[['Movie_Title', 'Rating', 'Revenue (Millions)']])
```

```python
# 14. Group by Director: avg Rating, Revenue
director_group = df.groupby('Director')[['Rating', 'Revenue (Millions)']].mean().sort_values(by='Rating', ascending=False)
print("Director Group Stats:\n", director_group)
```

```python
# 15. Group by Year: count, avg Rating
year_group = df.groupby('Year').agg({'Movie_Title': 'count', 'Rating': 'mean'}).rename(columns={'Movie_Title': 'Movie_Count', 'Rating': 'Avg_Rating'})
print("Movies per Year with Avg Rating:\n", year_group)
```

```python
# 16. Create Profitability column (Revenue / Duration)
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']

top_10_profitable = df.sort_values(by='Profitability', ascending=False).head(10)
top_10_profitable
```

```python
# 17. Categorize movies by Rating
def categorize_rating(r):
    if r < 6:
        return 'Poor'
    elif r < 7.5:
        return 'Average'
    else:
        return 'Excellent'

df['Rating_Category'] = df['Rating'].apply(categorize_rating)
```

```python
# 18. Custom scoring function
def custom_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(custom_score, axis=1)

# Optional: Display top scored movies
top_scored = df.sort_values(by='High_Score', ascending=False).head(10)
print("Top Scored Movies:\n", top_scored[['Movie_Title', 'High_Score']])
```
